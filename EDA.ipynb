>2024년 데이터는 제외하고 진행 : Nan 값이 2024년에만 있기 때문

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans 
```

```python
# Standard scale
target_df_ss = StandardScaler().fit_transform(target_df)
target_df_ss
```

```python
# 엘보우 방법을 사용하여 최적의 K값 탐색
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(target_df_ss)
    sse.append(kmeans.inertia_)

# SSE 그래프 그리기
plt.figure(figsize=(10, 6), facecolor='white')  # 배경을 흰색으로 설정
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

> k = 3개가 적당

```python
#2005~2023년 데이터 생성
label_data = data[:4955]
```

```python
model = KMeans(n_clusters=3,init='k-means++', max_iter=200, random_state=13)
cluster_lables = model.fit_predict(target_df_ss)
label_data['kmeans_label'] = cluster_lables
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
#실루엣평가

from sklearn.metrics import silhouette_samples, silhouette_score

silhouette_avg = silhouette_score(target_df_ss, cluster_lables)
score_values = silhouette_samples(target_df_ss, cluster_lables)

print('silhouette_avg',silhouette_avg)
```

```python
def correlation(num1, num2):
    temp = pd.concat([target, label_data.iloc[:, num1:num2]], axis=1)
    corr = temp.corr()
    corr= corr['USD']
    corr = corr.sort_values(ascending=False)
    return corr
```
